```python
def imputacionNumerica(variable, valor, df):
    df[variable].fillna(valor, inplace=True)

def imputar_columnas_numericas(df, dataSet='dfCasa_imputado.csv'):
    resumen = []

    columnas_irrelevantes = [col for col in df.columns if 'Unnamed' in col or df[col].isnull().all()]
    df.drop(columns=columnas_irrelevantes, inplace=True)
    print(f" Columnas irrelevantes eliminadas: {columnas_irrelevantes}")

    for col in df.columns:
        try:
        
            df[col] = pd.to_numeric(df[col], errors='raise')

            if df[col].notna().sum() > 0:
                promedio = df[col].mean()
                imputacionNumerica(col, promedio, df)
                resumen.append((col, promedio))
                print(f" Imputado '{col}' con promedio: {promedio}")
            else:
                print(f" Columna '{col}' solo contiene NaNs. No se imputa.")
        except:
            print(f" Columna '{col}' no es numérica. Se omite.")

    df.to_csv(dataSet, index=False)
    print(f"\n Archivo guardado como '{dataSet}'")


    if resumen:
        print("\n Resumen de imputación:")
        for col, valor in resumen:
            print(f" - {col}: imputado con {valor}")
    else:
        print("\n No se imputó ninguna columna.")

    return resumen
```

```python
def missing_cols(dfCasa):
    total = 0
    for col in dfCasa.columns:
        missing_vals = dfCasa[col].isnull().sum()
        total += missing_vals
        if missing_vals != 0:
            print(f'{col} => {dfCasa[col].isnull().sum()}')
    if total == 0:
        print('No hay valores faltantes')
 
missing_cols(dfCasa)
```
